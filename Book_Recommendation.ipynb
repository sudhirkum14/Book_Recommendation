# Popularity Based Recomender System

```python
rating_with_name = rating.merge(book,on='ISBN')
```

```python
num_rate_df=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rate_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rate_df
```

```python
avg_rate_df=rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rate_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rate_df
```

```python
numeric_cols = rating_with_name.select_dtypes(include='number').columns
avg_df = rating_with_name.groupby('Book-Title')[numeric_cols].mean().reset_index()
```

```python
avg_df.drop('User-ID',axis=1,inplace=True)
```

```python
avg_df
```

```python
popular_df=num_rate_df.merge(avg_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('Book-Rating',ascending=False).head(50)
popular_df = popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','Book-Rating']]
popular_df
```

# Collaborative Filtering Based Recommender System

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(padhe_likhe_users)]
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

    data=[]
    for i in similar_items:
        item=[]
        temp_df=book[book['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    return data
```

```python
recommend('1984')
```
